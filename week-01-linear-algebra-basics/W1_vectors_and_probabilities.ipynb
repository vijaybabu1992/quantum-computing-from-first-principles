```python
import numpy as np
```

```python
# Vector as States
state = np.array([0.6,0.8])
print("State Vector:",state)
```

```python
# Probability Amplitudes vs Probabilities
amplitudes = state

# Probabilites are squares of amplitudes
probabilities = amplitudes ** 2

print('Amplitudes:', amplitudes)
print("Probabilities:", probabilities)
print('Sum of probabilities:', probabilities.sum())

# Amplitudes can be any real numbers
# Squared amplitudes sum to 1
```

```python
# Why Normalization Matters

# Unnormalized vector
unnormalized = np.array([3, 4])

# Normalize the vector
norm = np.linalg.norm(unnormalized)
normalized = unnormalized / norm

print("Normalized vector:", normalized)
print("Sum of squared amplitudes:", (normalized ** 2).sum())

#Quantum states must always be normalized.
```
