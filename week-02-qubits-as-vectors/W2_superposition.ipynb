```python
import numpy as np
```

```python
# Basis states
ket_0 = np.array([1, 0])
ket_1 = np.array([0, 1])

print("|0> =", ket_0)
print("|1> =", ket_1)
```

```python
# Creating a Superposition

# Equal superposition
psi = (ket_0 + ket_1) / np.sqrt(2)

print("Superposition state |ψ>:", psi)
```

```python
# Measurement Probabilities
# Probabilities of measuring |0> and |1>
prob_0 = psi[0] ** 2
prob_1 = psi[1] ** 2

print("Probability of |0>:", prob_0)
print("Probability of |1>:", prob_1)
print("Total probability:", prob_0 + prob_1)
```

Superposition is just vector addition

Measurement probabilities come from squaring amplitudes

```python
# Measurement as Projection

# Projection onto |0> and |1>
proj_0 = np.dot(ket_0, psi)
proj_1 = np.dot(ket_1, psi)

print("Projection onto |0>:", proj_0)
print("Projection onto |1>:", proj_1)
```

This shows how measurement “selects” a basis direction.
