```python
import numpy as np
```

```python
# Define Qunatum States
psi = np.array([np.sqrt(0.3),np.sqrt(0.7)])

# Check normalization
print("State:", psi)
print("Sum of probabilities:", np.sum(psi**2))
```

```python
# Born Rule (Theoretical Probabilities)

probabilities = psi**2
print("Probability of |0>:", probabilities[0])
print("Probability of |1>:", probabilities[1])
```

```python
outcomes = [0,1]
measurement = np.random.choice(outcomes,p=probabilities)

print("Measured outcome:",measurement)
```

```python
num_trails = 10000
results = np.random.choice(outcomes, size=num_trails, p=probabilities)

count_0 = np.sum(results==0)
count_1 = np.sum(results==1)

print("Frequency |0>:",count_0/num_trails)
print("Frequency |1>:",count_1/num_trails)
```

```python
if measurement == 0:
    collapsed_state = np.array([1,0])
else:
    collapsed_state = np.array([0,1])

print("Collapsed state:", collapsed_state)
```
